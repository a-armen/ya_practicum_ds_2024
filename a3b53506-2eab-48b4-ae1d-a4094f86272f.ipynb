````python
def creator_pie(df, column_name, name, n, i):
    plt.subplot(1, n, i)
    plt.title(column_name)
    plt.pie(
        x=df.groupby(column_name)[name].count(), 
        startangle=90, 
        labeldistance=1.1, 
        wedgeprops={'linewidth':3, 'edgecolor':'white'}, 
        autopct='%1.1f%%', 
        labels=df.groupby(column_name)[name].count().index, 
        colors=sns.color_palette('BuPu_r')
    )
    
````

````python
def creator_hist(df, name_column, number, str_title, x_label, y_label):
    df[name_column].plot(kind='hist', grid=True, bins=number)
    plt.title(str_title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
    
def creator_boxplot(df, name_column, str_title, y_label, lim_x0, lim_x1):
    plt.boxplot(df[name_column])
    plt.title(str_title)
    plt.ylabel(y_label)
    plt.ylim(lim_x0, lim_x1)
    plt.grid()
    plt.show()
    
def create_barplot(x_aggfunc, df_pivot_table, color_name, title, label_x, label_y):
    f, ax = plt.subplots(figsize=(7,4))
    ax= sns.barplot(x=x_aggfunc, y=df_pivot_table.index,
                data=df_pivot_table, orient='h',
                palette=color_name)

    plt.title(title)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.xticks(rotation=45)
    plt.grid()
    sns.despine()
    plt.show()    
````

<font color='darkorange'>–¢–∞–∫ –∫–∞–∫ –º—ã —É–¥–∞–ª–∏–ª–∏ –ø—Ä–∏–∑–Ω–∞–∫ "id" —Å –Ω–µ–ø–æ–≤—Ç–æ—Ä—è—é—â–∏–º–∏—Å—è –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, —É –Ω–∞—Å –º–æ–≥—É—Ç –ø–æ—è–≤–∏—Ç—å—Å—è –Ω–æ–≤—ã–µ —è–≤–Ω—ã–µ –¥—É–±–ª–∏–∫–∞—Ç—ã: —á–µ–º –º–µ–Ω—å—à–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –æ—Å—Ç–∞—ë—Ç—Å—è, —Ç–µ–º –≤—ã—à–µ —à–∞–Ω—Å, —á—Ç–æ –∑–Ω–∞—á–µ–Ω–∏—è –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –±—É–¥—É—Ç –ø–µ—Ä–µ—Å–µ–∫–∞—Ç—å—Å—è. –ü—Ä–∏ —ç—Ç–æ–º —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –¥–ª—è –º–æ–¥–µ–ª–∏ —Ç–∞–∫–∏–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è —É–∂–µ –Ω–µ –ø—Ä–∏–Ω–µ—Å—É—Ç, –ø–æ—ç—Ç–æ–º—É —Å—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å –¥—É–±–ª–∏–∫–∞—Ç—ã —Ç–∞–∫–∂–µ –≤ —Ñ–∏–Ω–∞–ª—å–Ω–æ–π –≤–µ—Ä—Å–∏–∏ –æ–±—É—á–∞—é—â–µ–π —Ç–∞–±–ª–∏—Ü—ã.</font>

````python
RANDOM_STATE = 42

ohe_columns = ['dept', 'last_year_promo',  'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['supervisor_evaluation', 'employment_years', 'salary']

X_train = X_train_job_sat_rate.drop(['job_satisfaction_rate'], axis=1)
y_train = X_train_job_sat_rate['job_satisfaction_rate']
X_test = X_test_data.drop(['job_satisfaction_rate'], axis=1)
y_test = X_test_data['job_satisfaction_rate']
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è\
<font color='green'> –û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ —Ä–∞–∑–¥–µ–ª—è–µ—à—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –Ω–∞ –ø–æ—Ä—è–¥–∫–æ–≤—ã–µ –∏ –Ω–æ–º–∏–Ω–∞–ª—å–Ω—ã–µ.</font>

````python
# OneHotEncoder –∏ OrdinalEncoder –Ω–∞—Å—Ç—Ä–æ–∏–º –Ω–∞ —Ä–∞–±–æ—Ç—É —Å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–º–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏, –¥–ª—è —Ç–æ–≥–æ
# —á—Ç–æ–±—ã –∫–æ–¥ –Ω–µ —É–ø–∞–ª –µ—Å–ª–∏ –≤—Å—Ç—Ä–µ—Ç–∏—Ç –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã–µ –æ–Ω –Ω–µ –≤–∏–¥–µ–ª –Ω–∞ —ç—Ç–∞–ø–µ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.
ohe_pipe = Pipeline([
    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False))
])

ord_pipe = Pipeline([
    ('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'],
                    #['medium', 'high', 'low']
                    #['medium', 'low', 'high'],
                    ['low', 'medium', 'high']
                ],
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
])
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è\
<font color='green'> –ó–¥–æ—Ä–æ–≤–æ, —á—Ç–æ –µ—Å—Ç—å –æ–±—Ä–∞–±–æ—Ç–∫–∞ –Ω–µ–∏–∑–≤–µ—Å—Ç—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π.</font>

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚ùå\
<font color='red'> –û–±—Ä–∞—Ç–∏ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø–æ—Ä—è–¥–æ–∫ –∑–Ω–∞—á–µ–Ω–∏–π –≤ 'workload'</font>

<div style="background: #B0E0E6; padding: 5px; border: 1px solid SteelBlue; border-radius: 5px;">
    <font color='4682B4'><u><b>–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô –°–¢–£–î–ï–ù–¢–ê</b></u></font>
    <br />
    <font color='4682B4'>–ü—Ä–∏–º–µ–Ω–∏–ª –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ workload=['medium','low', 'high']</font>
</div>


<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ 2: </b></font> ‚ùå\
<font color='red'> –ü—Ä–∞–≤–∏–ª—å–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ ['low', 'medium', 'high'] ‚Üí ['–Ω–∏–∑–∫–∏–π', '—Å—Ä–µ–¥–Ω–∏–π', '–≤—ã—Å–æ–∫–∏–π']</font>

<div style="background: #B0E0E6; padding: 5px; border: 1px solid SteelBlue; border-radius: 5px;">
    <font color='4682B4'><u><b>–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô –°–¢–£–î–ï–ù–¢–ê V2</b></u></font>
    <br />
    <font color='4682B4'>–ò—Å–ø—Ä–∞–≤–∏–ª –Ω–∞ workload=['low', 'medium', 'high]</font>
</div>

````python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

# —Å–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LinearRegression())
])
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è\
<font color='green'> üëç</font>

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π

–û–±—É—á–∏—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–µ –º–æ–¥–µ–ª–∏. –í–æ–∑—å–º–∏—Ç–µ –æ–¥–Ω—É –ª–∏–Ω–µ–π–Ω—É—é –º–æ–¥–µ–ª—å, –∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –≤—Ç–æ—Ä–æ–π ‚Äî –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π. –ü–æ–¥–±–µ—Ä–∏—Ç–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –æ–¥–Ω–æ–π –º–æ–¥–µ–ª–∏ —Å –ø–æ–º–æ—â—å—é –æ–¥–Ω–æ–≥–æ –∏–∑ –∏–∑–≤–µ—Å—Ç–Ω—ã—Ö –≤–∞–º –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤.
–í—ã–±–µ—Ä–∏—Ç–µ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å –∏ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ. –í—ã–±–æ—Ä –¥–µ–ª–∞–π—Ç–µ –Ω–∞ –æ—Å–Ω–æ–≤–µ –Ω–æ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ ‚Äî SMAPE (–∞–Ω–≥–ª. symmetric mean absolute percentage error, ¬´—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∞–±—Å–æ–ª—é—Ç–Ω–æ–µ –ø—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ¬ª). 
–ú–µ—Ç—Ä–∏–∫–∞ SMAPE –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ç–∞–∫:

$$SMAPE = \frac{100}{n}*\sum_{i=1}^n\frac{|y_i-\widehat{y_i}|}{(|y_i| + |\widehat{y_i}|) / 2}$$

–≥–¥–µ:

- $y_i$ - —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–ª—è –æ–±—ä–µ–∫—Ç–∞ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º –Ω–æ–º–µ—Ä–æ–º i –≤ –≤—ã–±–æ—Ä–∫–µ;

- $\widehat{y_i}$ - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–ª—è –æ–±—ä–µ–∫—Ç–∞ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º –Ω–æ–º–µ—Ä–æ–º i –≤ –≤—ã–±–æ—Ä–∫–µ;
- `n` ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±—ä–µ–∫—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ;
- $\sum_{i=1}^n$ - —Å—É–º–º–∞ –∑–Ω–∞—á–µ–Ω–∏–π, –ø–æ–ª—É—á–µ–Ω–Ω–∞—è –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏–π, –∫–æ—Ç–æ—Ä—ã–µ —Å–ª–µ–¥—É—é—Ç –∑–∞ —ç—Ç–∏–º –∑–Ω–∞–∫–æ–º, –¥–ª—è –≤—Å–µ—Ö –æ–±—ä–µ–∫—Ç–æ–≤ —Å –ø–æ—Ä—è–¥–∫–æ–≤—ã–º –Ω–æ–º–µ—Ä–æ–º –æ—Ç i –¥–æ n –≤ –≤—ã–±–æ—Ä–∫–µ.



–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –º–∞—Å—Å–∏–≤—ã NumPy –∏–ª–∏ –æ–±—ä–µ–∫—Ç—ã Series –≤ pandas –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ SMAPE. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —ç—Ç—É –º–µ—Ç—Ä–∏–∫—É –ø—Ä–∏ –ø–æ–¥–±–æ—Ä–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ –æ—Ü–µ–Ω–∫–µ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π. 
–ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å–ø–µ—Ö–∞: SMAPE ‚â§ 15 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.



````python
def smape(y_true, y_pred):
    error = np.abs(y_true - y_pred)
    scale = (np.abs(y_true) + np.abs(y_pred)) / 2.0
    return np.mean(error / scale) * 100

smape_score = make_scorer(score_func=smape, greater_is_better=False)
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚úîÔ∏è\
<font color='green'> `smape` —Ä–∞—Å—á–∏—Ç–∞–Ω–∞ –≤–µ—Ä–Ω–æ.</font>

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è <br>
<font color='green'>–í–µ—Ä–Ω–æ, —á—Ç–æ –Ω–µ –∑–∞–±—ã–≤–∞–µ—à—å —É–∫–∞–∑–∞—Ç—å greater_is_better=False, –≠—Ç–æ –ø—Ä–∏–≥–æ–¥–∏—Ç—Å—è –¥–ª—è SearchCV. SMAPE ‚Äî –æ—à–∏–±–∫–∞, –µ—ë –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ–º.</font>

````python
randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring=smape_score,
    random_state=RANDOM_STATE,
    n_jobs=-1
).fit(X_train, y_train)
````

````python
def creator_pie_yes_no(df, column_name):
    plt.figure(figsize=(10, 3))
    plt.subplot(1, 2, 1)
    plt.title(f'–ü—Ä–∏–∑–Ω–∞–∫ "{column_name}": quit = yes')
    plt.pie(
        x=df[df['quit']=='yes'].groupby(column_name)['id'].count(), 
        startangle=90, 
        labeldistance=1.1, 
        wedgeprops={'linewidth':3, 'edgecolor':'white'}, 
        autopct='%1.1f%%', 
        labels=df.groupby(column_name)['id'].count().index, 
        colors=sns.color_palette('BuPu_r'))
    
    plt.subplot(1, 2, 2)
    plt.title(f'–ü—Ä–∏–∑–Ω–∞–∫ "{column_name}": quit = no')
    plt.pie(
        x=df[df['quit']!='yes'].groupby(column_name)['id'].count(), 
        startangle=90, 
        labeldistance=1.1, 
        wedgeprops={'linewidth':3, 'edgecolor':'white'}, 
        autopct='%1.1f%%', 
        labels=df.groupby(column_name)['id'].count().index, 
        colors=sns.color_palette('BuPu_r'))
    
````

### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

**–ü–æ–¥–≥–æ—Ç–æ–≤—å—Ç–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ç–∞–∫ –∂–µ, –∫–∞–∫ –∏ –≤ –ø–µ—Ä–≤–æ–π –∑–∞–¥–∞—á–µ: –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –ø–æ–¥–≥–æ—Ç–æ–≤–∫—É –≤ –ø–∞–π–ø–ª–∞–π–Ω–µ, –¥–æ–ø–æ–ª–Ω–∏–≤ –ø–∞–π–ø–ª–∞–π–Ω –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏. –ü—Ä–∏ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–∏ —É—á–∏—Ç—ã–≤–∞–π—Ç–µ –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –º–æ–¥–µ–ª–µ–π –∏ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–∞ –∫–æ–¥–∏—Ä–æ–≤—â–∏–∫–∞.**


````python
X_train_quit =  df_train_quit.set_index('id')
X_test_features_second = df_test_features_all.set_index('id')
y_test_target_quit = df_test_target_quit.set_index('id')

ohe_columns_new = ['dept', 'last_year_promo',  'last_year_violations',]
ord_columns_new = ['level', 'workload']
num_columns_new = ['supervisor_evaluation', 'employment_years', 'salary', 'job_satisfaction_rate']


X_train_second = X_train_quit.drop(['quit'], axis=1)
y_train_second = X_train_quit['quit']
X_test_second = X_test_features_second.drop(['quit'], axis=1)
y_test_second = y_test_target_quit['quit']
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ : </b></font> ‚úîÔ∏è\
<font color='green'> üëç</font>

````python
#OneHotEncoder –∏ OrdinalEncoder –Ω–∞—Å—Ç—Ä–æ–∏–º –Ω–∞ —Ä–∞–±–æ—Ç—É —Å –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–º–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏, –¥–ª—è —Ç–æ–≥–æ
#—á—Ç–æ–±—ã –∫–æ–¥ –Ω–µ —É–ø–∞–ª –µ—Å–ª–∏ –≤—Å—Ç—Ä–µ—Ç–∏—Ç –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ—Ç–æ—Ä—ã–µ –æ–Ω –Ω–µ –≤–∏–¥–µ–ª –Ω–∞ —ç—Ç–∞–ø–µ –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.
ohe_pipe_second = Pipeline([
    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False))
])

ord_pipe_second = Pipeline([
    ('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
    ('ord',  OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'],
                    #['medium', 'high', 'low']
                    #['medium', 'low', 'high']
                    ['low', 'medium', 'high']
                ],
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
])


param_grid_second = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 20),
        'models__max_features': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier()
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(10, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

# —Å–æ–∑–¥–∞—ë–º –æ–±—â–∏–π –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor_second = ColumnTransformer(
    [('ohe', ohe_pipe_second, ohe_columns_new),
     ('ord', ord_pipe_second, ord_columns_new),
     ('num', StandardScaler(), num_columns_new)
    ],
    remainder='passthrough'
)

# —Å–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω: –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –º–æ–¥–µ–ª—å
pipe_final_second = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚ùå\
<font color='red'> 'workload'</font>

<div style="background: #B0E0E6; padding: 5px; border: 1px solid SteelBlue; border-radius: 5px;">
    <font color='4682B4'><u><b>–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô –°–¢–£–î–ï–ù–¢–ê</b></u></font>
    <br />
    <font color='4682B4'>–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–ª –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ –≤ workload</font>
</div>


<div style="background: #B0E0E6; padding: 5px; border: 1px solid SteelBlue; border-radius: 5px;">
    <font color='4682B4'><u><b>–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô –°–¢–£–î–ï–ù–¢–ê V2</b></u></font>
    <br />
    <font color='4682B4'>–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–ª –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –Ω–∞ –ø–æ—Ä—è–¥–æ–∫ –≤ workload</font>
</div>

### –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

**–û–±—É—á–∏—Ç–µ –∫–∞–∫ –º–∏–Ω–∏–º—É–º —Ç—Ä–∏ –º–æ–¥–µ–ª–∏. –ö–∞–∫ –º–∏–Ω–∏–º—É–º –¥–ª—è –¥–≤—É—Ö –∏–∑ –Ω–∏—Ö –ø–æ–¥–±–µ—Ä–∏—Ç–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∫–∞—á–µ—Å—Ç–≤–æ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏. –ú–µ—Ç—Ä–∏–∫–∞ –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –≤ —ç—Ç–æ–π –∑–∞–¥–∞—á–µ ‚Äî ROC-AUC. –ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å–ø–µ—Ö–∞: ROC-AUC ‚â• 0.91 –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ. –ù–∞–ø–æ–º–Ω–∏–º: –æ—Ç–±–æ—Ä –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —á–∞—Å—Ç–æ –ø–æ–º–æ–≥–∞–µ—Ç —É–ª—É—á—à–∏—Ç—å –º–µ—Ç—Ä–∏–∫—É.**


````python
randomized_search_cv = RandomizedSearchCV(
    pipe_final_second,
    param_grid_second,
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
).fit(X_train_second, y_train_second)

print('–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å –∏ –µ—ë –ø–∞—Ä–∞–º–µ—Ç—Ä—ã:\n\n', randomized_search_cv.best_estimator_)
print ('–ú–µ—Ç—Ä–∏–∫–∞ –ª—É—á—à–µ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ:', randomized_search_cv.best_score_)

# –ø—Ä–æ–≤–µ—Ä—å—Ç–µ —Ä–∞–±–æ—Ç—É –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
# —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
y_test_pred_proba = randomized_search_cv.predict_proba(X_test_second)[:,1]
print(f'–ú–µ—Ç—Ä–∏–∫–∞ ROC-AUC –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ: {roc_auc_score(y_test_second, y_test_pred_proba)}')
````

<font color='blue'><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b></font> ‚ùå\
<font color='red'> ~~–ù—É–∂–Ω–æ –≤—ã–≤–µ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ü–µ–Ω–∫–∏ –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏, –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞.~~</font>

<div style="background: #B0E0E6; padding: 5px; border: 1px solid SteelBlue; border-radius: 5px;">
    <font color='4682B4'><u><b>–ö–û–ú–ú–ï–ù–¢–ê–†–ò–ô –°–¢–£–î–ï–ù–¢–ê</b></u></font>
    <br />
    <font color='4682B4'>–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –æ—Ü–µ–Ω–∫–∏ –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏</font>

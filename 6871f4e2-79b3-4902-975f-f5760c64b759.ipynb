```python
def add_separator_to_data_row(df_str):
    strings = df_str.split()
    new_string = '_'.join([word.lower() for word in strings])
    return new_string
    
```

```python
df_ml = df_ferma_main[df_ferma_main.columns]
```

```python
RANDOM_STATE = 42

X = df_ml.drop('удой', axis=1)
y = df_ml['удой'] 

# разобьем выбоки на тестовую и тренировочную
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)

# нормализуем категориальные признаки с помощью метода OneHotEncoder()

# категориальные признаки
columns_categorical_features = ['порода', 
                                'тип_пастбища', 
                                'возраст', 
                                'порода_папы_быка']

# количественные признаки
columns_quantitative_characteristics = ['энергетическая_кормовая_единица', 
                                        'сырой_протеин', 
                                        'сахаро_протеиновое_соотношение']

encoder = OneHotEncoder(drop='first', sparse_output=False)

X_train_ohe = encoder.fit_transform(X_train[columns_categorical_features])
X_test_ohe = encoder.transform(X_test[columns_categorical_features])

encoder_column_name = encoder.get_feature_names_out()

# Масштабируйте количественные признаки с помощью StandardScaler()
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train[columns_quantitative_characteristics])
X_test_scaler = scaler.transform(X_test[columns_quantitative_characteristics])


X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_column_name)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_column_name)

X_train_scaler = pd.DataFrame(X_train_scaler, columns=columns_quantitative_characteristics)
X_test_scaler = pd.DataFrame(X_test_scaler, columns=columns_quantitative_characteristics)

# объединим в тренировочный и тестовый датафреймы
X_train = pd.concat([X_train_ohe, X_train_scaler], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaler], axis=1)

# модель линейной регрессии
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
predictions = model_lr.predict(X_test)

# расчитаем коэффициент детерминации
r2 = r2_score(y_test, predictions)

print(f'R² LinearRegression: {r2:.3f}')
```

```python
RANDOM_STATE = 42

X = df.drop('вкус_молока', axis=1)
y = df['вкус_молока']


# разобьем выбоки на тестовую и тренировочную
X_train, X_test, y_train, y_test = train_test_split(
     X,
     y,
     random_state=RANDOM_STATE)

# # нормализуем категориальные признаки с помощью метода OneHotEncoder()

# категориальные признаки
columns_cat = ['порода', 
               'тип_пастбища',
               'порода_папы_быка',
               'возраст', 
               'имя_папы']

# количественные признаки
columns_num = \
    ['энергетическая_кормовая_единица', 
     'сырой_протеин', 
     'сахаро_протеиновое_соотношение', 
     'жирность', 
     'белок']


# категориальные признаки
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_train_ohe = encoder.fit_transform(X_train[columns_cat])
X_test_ohe = encoder.transform(X_test[columns_cat])

encoder_column_name = encoder.get_feature_names_out()

#количественные признаки
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train[columns_num])
X_test_scaler = scaler.transform(X_test[columns_num])


# объединение
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_column_name)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_column_name)

X_train_scaler = pd.DataFrame(X_train_scaler, columns=columns_num)
X_test_scaler = pd.DataFrame(X_test_scaler, columns=columns_num)

X_train = pd.concat([X_train_ohe, X_train_scaler], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaler], axis=1)

# обучение модели линейной регрессии и получение предсказаний на тестовых данных
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_prediction = clf.predict(X_test)

acc = accuracy_score(y_test, y_prediction)
recall = recall_score(y_test, y_prediction)
precision = precision_score(y_test, y_prediction)

print(f'accurancy = {acc:.3f}')
print(f'recall = {recall:.3f}')
print(f'precision = {precision:.3f}')
```
